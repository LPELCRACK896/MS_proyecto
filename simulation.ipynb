```python
from main import *
from VehiclesFactory import VehiclesFactory
import simpy
from CarStats import getMean_Desv
```

```python
cars_data = getMean_Desv("./car.csv")
print(cars_data)
```

```python
vehicle_factory = VehiclesFactory(
    name="global_factory",
    average_acceleration=cars_data["acceleration_mean"],
    de_acceleration=cars_data["acceleration_desv"],

    average_size=cars_data["length_mean"],
    de_size=cars_data["length_desv"]
    )
```

Establecemos el tiempo a simular en segundos
3600 segundos o una hora

```python
simulation_time = 3600
```

Creamos la calle de entrada
- 2 carros por segundo

```python
input_way = InputWay("START AUX PERIFERICO DERECHO", 0.2, vehicle_factory)
```

Simula la llegada de los carros a la calle por via de esta calle de entrada

```python
arrival_cars = input_way.calcular_tiempos_llegada(simulation_time)
```

Creamos la calle por la que transita de 904.4 metros con velocidad limite de 11m/s y una desviación estandar de 11.11

```python
road = Road("AUX PERIFERICO CARRIL DERECHO NORTE", 904.4, 11.11, 1.5)
```

```python
green_duration = 20  # Duración luz verde en segundos
red_duration = 40  # Duración luz roja en segundos
```

```python
env = simpy.Environment()

env.process(vehicle_arrival_process(env, input_way, road,  arrival_cars))
env.process(traffic_light_process(env, road, green_duration, red_duration))

env.run(until=simulation_time)
```
